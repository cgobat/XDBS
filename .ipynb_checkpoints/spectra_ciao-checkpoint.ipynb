```python
data = []
# model = 'mekal'
model = 'powerlaw'
# model = 'two mekal'
# model = 'bb'
model_dict = {'mekal': 'xsmekal.mek',
'powerlaw': 'xspowerlaw.pl',
'powerlawbb': '(xspowerlaw.pl + xsbbodyrad.bb)',
'two mekal': '(xsmekal.mek1 + xsmekal.mek2)',
'bb': 'xsbbodyrad.bb'}
# stat = 'chi2gehrels'
# stat = 'cstat'
stat = 'wstat'

# replace = False
replace = True

for i in range(1,int(nsources[0])+1):
# for i in range(43,44):
        try:
                #ui.set_xsxsect('vern')
                ui.set_stat(stat)
                print(name+'_src'+str(i)+'.pi')
                ui.load_pha(name+'_src'+str(i)+'.pi')
                ui.set_xlog()
                ui.set_ylog()
                ui.set_analysis('energy')
                ui.ignore(None, 0.5)
                ui.ignore(7, None)
                if stat == 'chi2gehrels':
                        ui.subtract()
                        ui.group_counts(5)
                else:
                        ui.unsubtract()
                        ui.group_counts(1)
                # ui.plot("data")
                # ui.ungroup()

                rsp = ui.get_response()
                ui.set_source('xsphabs.gal * '+model_dict[model])
                ui.xsphabs.gal.nH = 0.02
                if model == 'mekal':
                        ui.xsmekal.mek.kT = 0.3
                if model == 'two mekal':
                        ui.xsmekal.mek1.kT = 0.3
                        ui.xsmekal.mek2.kT = 2
                if model == 'powerlaw':
                        ui.xspowerlaw.pl.PhoIndex = 2.0
                if model == 'bb':
                        ui.xsbbodyrad.bb.kT = 0.3
                #ui.freeze(ui.xsphabs.gal.nH)
                ui.thaw(ui.xsphabs.gal.nH)
                # ui.set_bkg_model('gal*xsmekal.mek2')
                if model == 'mekal':
                        ui.guess('mek')
                if model == 'two mekal':
                        ui.guess('mek1')
                        ui.guess('mek2')
                if model == 'powerlaw':
                        ui.guess('pl')
                if model == 'bb':
                        ui.guess('bb')
                #ui.show_source()
                ui.fit()
                ui.show_model()

                # ui.set_bkg_model('gal*xsmekal.mek2')

                if stat == 'chi2gehrels':
                        # # calculate goodness of fit chi2
                        ui.calc_stat_info()
                        rstat = ui.get_stat_info()[0].rstat
                else:
                        # calculate goodness of fit cstat
                        ui.calc_stat_info()
                        obscstat = ui.calc_stat()
                        qval = ui.get_stat_info()[0].qval
                        rstat = ui.get_stat_info()[0].rstat
                        print("observed cstat = ", obscstat)
                        x = ui.get_data_plot().x
                        y = ui.get_data_plot().y
                        [Ce,Cv] = cstat_gof.gof_cstat(x,y)
                        print("expected cstat = ",Ce," +- ",np.sqrt(Cv))
                        qual = (obscstat-Ce)/np.sqrt(Cv)
                        print("Observed cstat is off from expected by ",qual," sigma")

                print("FLUX AND COUNTS")
                flux=ui.calc_energy_flux(0.5, 7.0)
                counts=ui.calc_data_sum(0.5, 7.0)
                # set analysis to energy with factor one such that count rate is in counts/s
                ui.set_analysis('energy', factor=0)
                # count_rate=ui.get_rate()
                print(flux)
                print(counts)
                # print(np.array(count_rate).mean())
                ui.get_conf().max_rstat = 1e30
                ui.conf()
                # ui.reg_proj('pl.PhoIndex', 'gal.nH')

                # save source properties, parameter values, and fitted parameter errors into list
                if stat=='chi2gehrels':
                        data.append([name+'_src'+str(i), flux, counts, rstat] + [par.val for par in ui.get_model().pars] + list(np.sqrt(ui.get_fit_results().covar.diagonal())+ list(ui.get_conf_results().parmins) + list(ui.get_conf_results().parmaxes))) 
                else:
                        data.append([name+'_src'+str(i), flux, counts, obscstat, Ce, Cv, qual, qval, rstat] + [par.val for par in ui.get_model().pars] + list(np.sqrt(ui.get_fit_results().covar.diagonal())) + list(ui.get_conf_results().parmins) + list(ui.get_conf_results().parmaxes)) 

                if replace == True:
                        ui.ignore(None, 0.5)
                        ui.ignore(7, None)
                        ui.subtract()
                        ui.group_counts(15)
                        wstat_model = ui.get_fit_plot().modelplot
                        # refit with chi2gehrels, take data from that fit, and plot
                        ui.set_stat('chi2gehrels')
                        ui.subtract()
                        ui.fit()
                        ui.show_model()
                        
                        fig = plt.figure(figsize=(6, 4))
                        plt.rcParams.update({'font.size': 16})
                        gs = fig.add_gridspec(2, 1, hspace=0.05,height_ratios=[3, 1])

                        ax1 = fig.add_subplot(gs[0])
                        plt.sca(ax1)
                        ui.plot_data(xlog=True, ylog=True, overplot=True, clearwindow=False,)
                        wstat_model.plot(xlog=True, ylog=True, overplot=True, clearwindow=False,)
                        plt.xlim([0.5,8.0])
                        plt.xscale('log')
                        plt.yscale('log')
                        plt.gcf().axes[0].set_title(f'Source {i} {model.title()} Fit')
                        ax1.set_ylabel('Counts/s/keV')


                        ax2 = fig.add_subplot(gs[1], sharex=ax1)
                        plt.sca(ax2)

                        # exactly replicates plot_ratio
                        plt.errorbar(x=ui.get_fit_plot().dataplot.x, y=ui.get_fit_plot().dataplot.y/wstat_model.y, xerr=ui.get_fit_plot().dataplot.xerr/2, yerr=ui.get_fit_plot().dataplot.yerr/wstat_model.y, fmt='.')
                        plt.yscale('log')
                        ax2.get_xaxis().set_minor_formatter(matplotlib.ticker.StrMethodFormatter('{x:,g}'))
                        ax2.get_xaxis().set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,g}'))
                        ax2.set_xticks([0.6, 1, 2, 3, 4, 5, 6, 7], minor=True)
                        ax2.get_yaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
                        ax2.get_yaxis().set_minor_formatter(matplotlib.ticker.ScalarFormatter())
                        ax2.set_yticks([], minor=True)
                        ax2.ticklabel_format(style='plain', axis='y')
                        ax2.set_xlabel('Energy (keV)')
                        ax2.set_ylabel('Data/Model')


                elif replace == False:
                        ui.group_counts(15)
                        print('test')
                        if stat=='chi2gehrels':
                                ui.plot_fit_delchi(xlog=True, ylog=True)
                        else:
                                ui.plot_fit_ratio(xlog=True, ylog=True)
                        plt.xlim([0.5,8.0])
                        plt.gcf().axes[0].set_title(f'Source {i} {model.title()} Fit')
                        plt.gcf().axes[1].get_xaxis().set_minor_formatter(matplotlib.ticker.StrMethodFormatter('{x:,g}'))
                        plt.gcf().axes[1].get_xaxis().set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,g}'))                        
                        plt.gcf().axes[1].set_xticks([0.5, 1, 2, 3, 4, 5, 6, 7], minor=True)
                        plt.gcf().axes[1].get_yaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
                        plt.gcf().axes[1].get_yaxis().set_minor_formatter(matplotlib.ticker.ScalarFormatter())
                        plt.gcf().axes[1].set_yticks([], minor=True)
                        plt.gcf().axes[1].ticklabel_format(style='plain', axis='y')
                        plt.gcf().axes[1].set_ylabel('Residuals')

                plt.savefig((f'{name}_src_{i}_{model}.png').replace(' ','_'), dpi=300)
                ui.ungroup()
                ui.reset()
        except Exception as e:
                # not necessarily the same size as sources that were fit, still works
                data.append([name+'_src'+str(i), flux, counts, None] + [None, None, None] + [None, None, None])
                print("Error: " + str(e))
```

```python
# for each source, find flare start and end times, if any
bins = 50

lines = open('spectra.lis', 'r')
df_flares = pd.DataFrame(index=range(117), columns=['times', 'median', 'start', 'end', 'duration'])
df_flares.index = df_flares.index+1
for i, line in enumerate(lines):
    line = line.strip()
    line = line.replace('(', '\(')
    line = line.replace(')', '\)')
    line += "[cols\ time]"
    # print(line)
    times = !dmlist $line opt=data,clean
    times = np.array(list(map(str.strip, times[1:]))).astype(float)


    # bin the data and get the median of the binned data
    binned_times = np.histogram(times, bins=bins)
    median = np.median(binned_times[0])

    # set median to 1 if 0
    if median == 0:
        median = 1

    # for each bin, get the poisson probability that the bin counts didn't occur by chance
    probs = poisson.cdf(binned_times[0], median)

    df_flares.iloc[i]['median'] = median
    df_flares.iloc[i]['times'] = times

    # find first bin with probs 4.5 sigma above the median to indicate start of flare, and the next bin with probs 3 sigma below the median to indicate end of flare
    start = np.nan
    end = np.nan
    for j, prob in enumerate(probs):
        if np.isnan(start):
            if prob >= 0.99999320:
                start = binned_times[1][j]
        else:
            if prob < 0.99:
                end = binned_times[1][j]
                df_flares.iloc[i][['start', 'end', 'duration']] = [start, end, end-start]
                break
    
    # if no end is found, set end to the last time in the flare
    if ~np.isnan(start):
        if np.isnan(end):
            end = binned_times[1][-1]
            df_flares.iloc[i][['start', 'end', 'duration']] = [start, end, end-start]

    
```
